## <font color='264CC7'> Introducción a la Representación de Funciones</font>
---

En este notebook, se calcula la diferentes formas de representar una función a partir de su representación algebraica:
1. Representación gráfica
2. Representación tabular


```python
#@title <font color="264CC7" size="+1"> Ejecutar configuración
# Importamos las bibliotecas necesarias
import pandas as pd
from sympy import *
```

```python
# @title ## <font color='264CC7'> Representación Gráfica</font> {"form-width":"50%","display-mode":"form"}

variable = 'x' # @param {type:"string"}
expression = 'x**2 + 2*x + 1' # @param {type:"string"}
x_min = -2 # @param {type:"number"}
x_max = 4 # @param {type:"number"}
y_min = 0 # @param {type:"number"}
y_max = 20 # @param {type:"number"}

# Definimos la variable simbólica
x = symbols(variable)

# Definimos la función
f = Lambda(x, expression)

print(f"La función definida es: f({variable}) = ", f(x))

plot(f(x), (x,x_min,x_max), ylim=(y_min,y_max))
```

```python
# @title <font color='264CC7'> Tabla de Valores</font> {"form-width":"50%","display-mode":"form"}

variable = 'x' # @param {type:"string"}
expression = 'x**2 + 2*x + 1' # @param {type:"string"}
x_min = -5 # @param {type:"number"}
x_max = 5 # @param {type:"number"}
d_x = 1 # @param {type:"number"}

# Definimos la variable simbólica
x = symbols(variable)

# Definimos la función
f = Lambda(x, expression)

print(f"La función definida es: f({variable}) = ", f(x))

# Generamos algunos valores de x
x_valores = range(x_min, x_max+d_x, d_x)

# Calculamos los valores correspondientes de y
y_valores = [f(x) for x in x_valores]

# Creamos un DataFrame de pandas
df = pd.DataFrame({f'{variable}': x_valores, f'f({variable})': y_valores})

# Mostramos la tabla
display(df)
```
